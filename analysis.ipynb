### Analysis of Football season and team stats

```python
# load the csv file created after scraping the data from 'data' folder of the project
import pandas as pd

df = pd.read_csv(r'./data/df.csv')
df.head()
```

```python
pd.options.display.max_columns = None
pd.options.display.max_rows = None
```

```python
def change_into_datetime(col):
    df[col]=pd.to_datetime(df[col])
change_into_datetime("Date")
df.head()
```

```python
#removing betting data columns from original df
#removing 21-22 season as it is incomplete and not suitable for further analysis

df_complete = df.iloc[323: , :24] 
df_complete = df_complete.sort_index(ascending=False)

df_complete =df_complete.drop(['Div','Time'],axis=1)
df_complete.head()
```

```python
df_complete = df_complete.dropna()
# df_complete.isnull().sum()
```

Drop irrelevant columns and rename other columns to understand the contents from column title

```python
df_complete.columns = ['Date','HomeTeam','AwayTeam','FT_Home_Goal','FT_Away_Goal','FT_Result','HT_Home_Goal','HT_Away_Goal',
                        'HT_Result','Referee','H_Shots','A_Shots','H_Shots_Target','A_Shots_Target','H_Foul',
                        'A_Foul','H_Corner','A_Corner','H_Yellow','A_Yellow','H_Red','A_Red']

df_complete.head()
```

```python
#sorting data according to the Date

df_analysis = df_complete.sort_values(['Date']).reset_index(drop=True)
df_analysis.head()
```

Each EPL season has 380 games.
So we can make a new column called season which changes after every 380 games.

```python
season_start=2
season_end=3
season_list = []

for x in range (18):
    for y in range (380):
        season_list.append((str(season_start)+'-'+str(season_end)))
    season_start = season_start + 1
    season_end = season_end + 1

# season_list
```

In the above step, we made an empty list and added the seasons into it changing every 380 games.
We will convert this list into a pandas dataframe and later concatenate this data frame with the df_analysis.

```python
season_df = pd.DataFrame({'Season':season_list})

data_by_seasons = pd.concat([season_df,df_analysis], axis=1)
data_by_seasons.head()
```

```python
data_by_seasons = data_by_seasons.dropna()
```
