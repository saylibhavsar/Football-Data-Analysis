### Analysis of Football season and team stats

```python
# load the csv file created after scraping the data from 'data' folder of the project
import pandas as pd

df = pd.read_csv(r'./data/df.csv')
df.head()
```

```python
pd.options.display.max_columns = None
```

```python
pd.options.display.max_rows = None
```

```python
def change_into_datetime(col):
    df[col]=pd.to_datetime(df[col])
change_into_datetime("Date")
df.head()
```

```python
#removing betting data columns from original df
#removing 21-22 season as it is incomplete and not suitable for further analysis

df_complete = df.iloc[323: , :24] 
df_complete = df_complete.sort_index(ascending=False)

df_complete =df_complete.drop(['Div','Time'],axis=1)
df_complete.head()
```

```python
df_complete = df_complete.dropna()
# df_complete.isnull().sum()
```

Drop irrelevant columns and rename other columns to understand the contents from column title

```python
df_complete.columns = ['Date','HomeTeam','AwayTeam','H_Goals','A_Goals','FT_Result','H_HT_Goals','A_HT_Goals',
                        'HT_Result','Referee','H_Shots','A_Shots','H_Shots_Target','A_Shots_Target','H_Foul',
                        'A_Foul','H_Corner','A_Corner','H_Yellow','A_Yellow','H_Red','A_Red']

df_complete.head()
```

```python
#sorting data according to the Date

df_analysis = df_complete.sort_values(['Date']).reset_index(drop=True)
df_analysis.head()
```

Each EPL season has 380 games.
So we can make a new column called season which changes after every 380 games.

```python
season_start=2
season_end=3
season_list = []

for x in range (18):
    for y in range (380):
        season_list.append((str(season_start)+'-'+str(season_end)))
    season_start = season_start + 1
    season_end = season_end + 1

# season_list
```

In the above step, we made an empty list and added the seasons into it changing every 380 games.
We will convert this list into a pandas dataframe and later concatenate this data frame with the df_analysis.

```python
season_df = pd.DataFrame({'Season':season_list})

data_by_seasons = pd.concat([season_df,df_analysis], axis=1)
data_by_seasons.head()
```

```python
data_by_seasons = data_by_seasons.dropna()
data_by_seasons.head()
```

We need to stack HomeTeam and AwayTeam into one column called Team - to get season & team wise data

```python
delta_names = ['Goals','HT_Goals','Shots_Target','Foul','Corner','Yellow','Red']

for column in delta_names:
    h_delta_column = 'H_Delta_'+ column
    a_delta_column = 'A_Delta_'+ column
    
    h_column = 'H_'+ column
    a_column = 'A_'+ column

    data_by_seasons[h_delta_column] = data_by_seasons[h_column]-data_by_seasons[a_column]
    data_by_seasons[a_delta_column] = data_by_seasons[a_column]-data_by_seasons[h_column]
    
column_names = ['Delta_Goals','Delta_HT_Goals','Delta_Shots_Target','Delta_Foul','Delta_Corner','Delta_Yellow','Delta_Red']

h_column_names = ['Season','HomeTeam','H_Goals','A_Goals','FT_Result','H_HT_Goals','A_HT_Goals','HT_Result','H_Shots_Target','A_Shots_Target','H_Foul','A_Foul','H_Corner','A_Corner','H_Yellow','A_Yellow','H_Red','A_Red']

a_column_names = ['Season','AwayTeam','A_Goals','H_Goals','FT_Result','A_HT_Goals','A_HT_Goals','HT_Result','A_Shots_Target','H_Shots_Target','A_Foul','H_Foul','A_Corner','H_Corner','A_Yellow','H_Yellow','A_Red','H_Red']


column_names_new = ['Season','Team','Goals','Goals_Received','FT_Result','HT_Goals','HT_Goals_Received','HT_Result','Shots_Target','Shots_Target_Received','Fouls','Got_Fouled','Corner','Corner_Received','Yellow','Yellow_Received','Red','Red_Received','Delta_Goals','Delta_HT_Goals','Delta_Shots_On_Target','Delta_Foul','Delta_Corner','Delta_Yellow','Delta_Red']

for column in column_names: 
    h_column_names.append("H_" + column)
    a_column_names.append("A_" + column)
```

```python
df_home = data_by_seasons.filter(h_column_names)
df_away = data_by_seasons.filter(a_column_names)

df_home.columns = column_names_new
df_away.columns = column_names_new

# df_total = df_home.append(df_away, ignore_index=True).sort_values(['Season','FT_Result', 'HT_Result'], ascending=[True,True, True])

# df_total_sorted = df_total['Season','Team','Goals','Goals_Received','Delta_Goals','FT_Result','HT_Goals','HT_Goals_Received','Delta_HT_Goals','HT_Result','Shots_On_Target','Shots_On_Target_Received','Delta_Shots_On_Target','Fouls','Got_Fouled','Delta_Fouls','Corners','Corners_Received','Delta_Corners','Yellow','Yellow_Received','Delta_Yellow','Red','Red_Received','Delta_Red']

df_total = df_home.append(df_away)
df_total.head()
```
